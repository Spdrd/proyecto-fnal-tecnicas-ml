```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configure visualization style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Load the data
# Make sure 'health_data.csv' is in the same folder as your notebook
file_path = 'health_data.csv'
df = pd.read_csv(file_path)

# Display the first few rows to verify loading
display(df.head())
```

```python
# General information about columns and data types
print("Dataset Information:")
print(df.info())

# Dataset dimensions
print(f"\nShape of the dataset: {df.shape}")
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values[missing_values > 0])

# Check for duplicates
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
```

```python
# Descriptive statistics for numerical variables
display(df.describe())
```

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Plot histograms
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Plot boxplots for numerical variables
for col in numerical_cols:
    plt.figure(figsize=(8, 2))
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# Select categorical columns (type 'object')
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    # Calculate value counts and percentages
    counts = df[col].value_counts()
    percentages = df[col].value_counts(normalize=True) * 100
    
    # Create a dataframe for better display
    summary = pd.DataFrame({'Count': counts, 'Percentage (%)': percentages})
    print(f"\n--- Summary for {col} ---")
    display(summary)
    
    # Plot bar chart
    plt.figure(figsize=(8, 4))
    ax = sns.barplot(x=counts.index, y=counts.values, palette="viridis")
    plt.title(f'Distribution of {col}')
    plt.ylabel('Count')
    plt.xlabel(col)
    plt.xticks(rotation=45) # Rotate labels if they are long
    
    # Add percentage labels on top of bars
    for i, p in enumerate(ax.patches):
        percentage = f'{percentages[i]:.1f}%'
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='bottom')
        
    plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

# Transformaciones

```python
import pandas as pd

# Diccionario con medias y desviaciones estándar proporcionadas
stats = {
    "Enfermedad cardiovascular": {"mean": 0.303130, "std": 0.514926},
    "Diabetes": {"mean": 0.205536, "std": 0.450558},
    "Asma": {"mean": 0.150921, "std": 0.399708},
    "Cáncer": {"mean": 0.105589, "std": 0.342888},
    "Obesidad": {"mean": 0.248486, "std": 0.483748},
    "Depresión/Ansiedad": {"mean": 0.401899, "std": 0.545406}
}

# 1. Desestandarizar
for col in stats:
    df[col + "_real"] = df[col] * stats[col]["std"] + stats[col]["mean"]

# 2. Convertir a binario (0 = no, 1 = sí)
for col in stats:
    df[col + "_bin"] = (df[col + "_real"] >= 0.5).astype(int)

    df.to_csv('health_data_processed.csv', index=False)

df.head()
```

```python
disease_cols = [
    "Enfermedad cardiovascular_real",
    "Diabetes_real",
    "Asma_real",
    "Cáncer_real",
    "Obesidad_real",
    "Depresión/Ansiedad_real"
]

df_diseases = df[disease_cols]
```
